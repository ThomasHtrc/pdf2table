## Extract from PDF

```python
# Initialize Driver for processing
from pdf2table import Driver
driver = Driver()
```

```python
table_dict = driver.extract_tables('test_data/test_document.pdf', implicit_rows=True, borderless_tables=True)
```

```python
from IPython.display import HTML
import pandas as pd

for page_num, table_list in table_dict.items():
    print(f'{page_num}')
    for tab in table_list:
        display(HTML(tab.to_html()))
```
